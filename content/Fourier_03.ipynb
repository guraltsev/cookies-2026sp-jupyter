# Activity: Two ways to write the same sinusoid

This activity compares two common ways to write a 1‑periodic sinusoid:

1. **Linear combination form**
$$
f(x)=a\sin(2\pi x)+b\cos(2\pi x).
$$

2. **Amplitude–phase form**
$$
g(x)=C\sin(2\pi x-\phi).
$$

Your goal is to experiment numerically and discover how $(a,b)$ corresponds to $(C,\phi)$.
